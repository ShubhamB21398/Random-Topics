```python
sentance = 'Start a sentence and then bring it to an end'
```

```python
pattern = re.compile(r'abc')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
# One match found for 'abc' from staring index 1 to 4
```

```python
pattern = re.compile(r'cba')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
# Found no match for 'cba'
```

```python
pattern = re.compile(r'.')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
# Since, period is a special character, so it will select every character if we didn't used escape character before it.
# Hence, it is giving every character as result.
```

```python
pattern = re.compile(r'\.')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
# Here, we have used escape character before period, hence we are getting matches for '.' only.
```

```python
#.   Any character except new line
pattern = re.compile(r'.')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
# Here, we used '\.' to macth period.
```

```python
# Function to match raw text within test string.
def match_regex(raw_text, TextToSearch):
    pattern = re.compile(raw_text)
    matches = pattern.finditer(TextToSearch)
    for match in matches:
        print(match)
```

```python
sites = '''
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
'''
```

```python
# Grouping
pattern = re.compile(r'https?://(www.)?(\w+)\.(\w+)')
matches = pattern.finditer(sites)
for match in matches:
    print(match.group(0)) # Original string
    print(match.group(1)) # First group
    print(match.group(2)) # Second group
    print(match.group(3)) # Third group
```

```python
pattern = re.compile(r'https?://(www.)?(\w+)(\.\w+)')
subbed_urls = pattern.sub(r'\2\3',sites)
print(subbed_urls)
```
